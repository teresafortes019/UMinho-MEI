Fórmulas Proposicionais:

    3. CPU2 -> MB2: O CPU2 só pode ser instalado na motherboard MB2.
    4. MON1 -> PG1 ^ RAM2: O monitor MON1 para poder funcionar precisa da placa gráfica PG1 e da memória RAM2.
    5. MON2 ^ PG3 -> RAM2: O monitor MON2 precisa da memória RAM2 para poder trabalhar com a placa gráfica PG3.
    6. (CP1 ^ ¬CP2) v (¬CP1 ^ CP2): O computador tem um único CPU.
    7. (RAM1 ^ ¬RAM2) v (¬RAM1 ^ RAM2): O computador tem uma única memória RAM. 
    8. (MB1 ^ ¬MB2) v (¬MB1 ^ MB2): O computador tem uma única Motherboard.
    9. (PG1 V PG2 V PG3) ∧ (¬PG1 ∨ ¬PG2 ∨ ¬PG3): O computador tem uma única Placa Gráfica.
    10. MON1 V MON2 V MON3: O computador poderá ter ou não ter monitores.

    9.  (PG1 V PG2 V PG3) ∧ (¬PG1 ∨ ¬PG2 ∨ ¬PG3)
